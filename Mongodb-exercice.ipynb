```python
import json
import pymongo
import pandas as pd
import json
myclient = pymongo.MongoClient()

df = pd.read_csv('films.csv',encoding = 'ISO-8859-1')   # loading csv file

def mongoimport(csv_path, db_name, coll_name, db_url='localhost', db_port=27000):
    """ Imports a csv file at path csv_name to a mongo colection
    returns: count of the documants in the new collection
    """
    client = myclient
    db = client[db_name]
    coll = db[coll_name]
    data = pd.read_csv(csv_path)
    payload = json.loads(data.to_json(orient='records'))
    coll.remove()
    coll.insert(payload)
    return coll.count()
    
mongoimport("movie_data.csv", "imdb", "films")
mongoimport("serie_crawl.csv", "imdb", "series")


```
